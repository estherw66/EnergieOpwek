# Cleaning windmills data

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
```

## Loading data 

```python
df = pd.read_csv('../data/Other/Windmolens_2021-2023.csv',sep=';', skiprows=3)
df
```

## Cleaning data

So the data is not 

```python
# Rename columns for simplicity and readability
df.columns = [
    "Region",
    "Wind Turbines 2021", "Wind Turbines 2022", "Wind Turbines 2023",
    "Capacity MW 2021", "Capacity MW 2022", "Capacity MW 2023",
    "Production kWh 2021", "Production kWh 2022", "Production kWh 2023",
    "Capacity Factor 2021 (%)", "Capacity Factor 2022 (%)", "Capacity Factor 2023 (%)"
]

# Display cleaned DataFrame
df
```

```python
# Drop the first two rows
df = df.drop([0, 1,15]).reset_index(drop=True)
df
```

```python
# Melt the DataFrame to long format for each year
df_melted = pd.melt(df, 
                    id_vars=["Region"], 
                    value_vars=[
                        "Wind Turbines 2021", "Wind Turbines 2022", "Wind Turbines 2023",
                        "Capacity MW 2021", "Capacity MW 2022", "Capacity MW 2023",
                        "Production kWh 2021", "Production kWh 2022", "Production kWh 2023",
                        "Capacity Factor 2021 (%)", "Capacity Factor 2022 (%)", "Capacity Factor 2023 (%)"
                    ],
                    var_name="Metric and Year",
                    value_name="Value")

df_melted
```

```python
# Extract the year and metric into separate columns
df_melted["Year"] = df_melted["Metric and Year"].str.extract(r'(\d{4})')
df_melted["Metric"] = df_melted["Metric and Year"].str.extract(r'^(.*?) \d{4}')

# Ensure numeric columns are clean
df_melted["Value"] = pd.to_numeric(df_melted["Value"].str.replace(",", "."), errors="coerce")

df_melted
```

```python
# Pivot to have metrics as columns
df_final = df_melted.pivot_table(index=["Region", "Year"], 
                                 columns="Metric", 
                                 values="Value").reset_index()
df_final
```

```python
# Clean up column names
df_final.columns.name = None
df = df_final.rename(columns={
    "Wind Turbines": "Wind Turbines",
    "Capacity MW": "Capacity (MW)",
    "Production kWh": "Production (kWh)",
    "Capacity Factor (%)": "Capacity Factor (%)"
})

df
```

```python
df.to_csv('../data/Other/Cleaned_Windmolens_2021-2023.csv', index=False)
```
