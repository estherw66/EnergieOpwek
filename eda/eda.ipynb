```python
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
df= pd.read_csv('../data/ProductionData/energie_opwek.csv')
df
```

```python
df['type'].value_counts().plot(kind='bar')
```

```python
df['activity'].value_counts().plot(kind='bar')
```

```python
df['classification'].value_counts().plot(kind='bar')
```

```python
# Sort by Type (optional, for better visualization)
df = df.sort_values(by='type')

# Plot the line graph
plt.plot(df['type'], df['emissionfactor (kg CO2/kWh)'], marker='o', color='b', label='Emission Factor')

# Add labels and title
plt.title('Emission Factors by Type')
plt.xlabel('Type')
plt.ylabel('Emission Factor')
plt.legend()

# Show the plot
plt.show()
```
